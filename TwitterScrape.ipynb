```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import itertools
import numpy as np
```

```python
# create list to append tweet data to
tweets_list = []

# create scraper function for selected search terms using TwitterSearchScraper
def tweet_scraper(searches, since_date, until_date):
    for x in searches:
        for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f'{x} since:{since_date} until:{until_date}').get_items()):
            tweets_list.append([tweet.id, tweet.date, tweet.content, tweet.user.username, tweet.user.location, tweet.user.id])
            if i>2:
                 break
        
# select 10 neutral search terms
# search syntax documentation: https://github.com/igorbrigadir/twitter-advanced-search
terms = ['#maskmandate',
         'masks are *',
         '#masks',
         '#mask',
         '(mask public)',
         '(mask mandate)',
         '(mask CDC)',
         '(wearing a mask is *)',
         '(mask children) OR (mask kids)',
         '(mask government)']
```

```python
tweet_scraper(terms, '2020-05-01', '2020-05-31')
tweet_scraper(terms, '2020-06-01', '2020-06-30')
tweet_scraper(terms, '2020-07-01', '2020-07-31')
tweet_scraper(terms, '2020-08-01', '2020-08-31')
# create a dataframe from the tweets list above
tweeter_df = pd.DataFrame(tweets_list, columns=['tweet.id', 'tweet.date', 'content', 'tweet.username', 'user.location', 'user.id'])
tweeter_df
```
