```python
csv_file = "tweets_final.csv"
tweets_final_df = pd.read_csv(csv_file)
tweets_final_df.head()
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5432/Mask_Data')
conn = engine.connect()
```

```python
engine.table_names()
```

```python
apr_data_df.to_sql(name='april', con=engine, if_exists='append', index=False)
may_data_df.to_sql(name='may', con=engine, if_exists='append', index=False)
jun_data_df.to_sql(name='june', con=engine, if_exists='append', index=False)
july_data_df.to_sql(name='july', con=engine, if_exists='append', index=False)
aug_data_df.to_sql(name='august', con=engine, if_exists='append', index=False)
sep_data_df.to_sql(name='september', con=engine, if_exists='append', index=False)
oct_data_df.to_sql(name='october', con=engine, if_exists='append', index=False)




```

```python
tweets_final_df.to_sql(name='tweets', con=engine, if_exists='append', index=False)
```
