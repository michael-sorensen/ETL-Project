```python
import nltk.sentiment.vader
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
pos_list = []
neg_list = []
neu_list = []
comp_list = []

for sentence in tweets_df['content']:
    vs = analyzer.polarity_scores(sentence)
    pos_list.append(vs['pos'])
    neg_list.append(vs['neg'])
    neu_list.append(vs['neu'])
    comp_list.append(vs['compound'])
    
tweets_df['Positive'] = pos_list
tweets_df['Negative'] = neg_list
tweets_df['Neutral'] = neu_list
tweets_df['Compound'] = comp_list
tweets_df.sort_values(by=['Compound'])
```

```python
tweets_df =  tweets_df[tweets_df.Compound !=0]
tweets_df
```
